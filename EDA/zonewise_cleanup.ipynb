```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./extracted/zonewise_iter1.csv')
df.head(10)
```

```python
df = df.drop(columns=['File Name'])
df.head(10)
```

```python
df['Report Date'] = pd.to_datetime(df['Report Date'], format='%d.%m.%y')
df.head(10)
```

```python
df = df.sort_values(by='Report Date', ascending=False).reset_index(drop=True)
df.head(9)
```

```python
df['Generation(MW)'] = df['Probable Day Peak (MW)'] + df['Probable Evening Peak (MW)']
df['Generation(MW)'] = df['Generation(MW)'].round().astype(int)
df.head(10)
```

```python
df = df.drop(columns=['Probable Day Peak (MW)', 'Probable Evening Peak (MW)'])
df.head(10)
```
