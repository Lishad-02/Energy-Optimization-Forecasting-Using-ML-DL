```python
# Histogram for Max Demand
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['MaxDemand(MW)'], bins=30, kde=True)  # kde=True adds a smooth density curve
plt.xlabel('Max Demand (MW)')
plt.ylabel('Count')
plt.title('Histogram of Max Demand with KDE')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot boxplots for outlier detection
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Temperature'])
plt.title('Boxplot of Temperature')


plt.show()
```

```python


Q1 = df["Temperature"].quantile(0.25)
Q3 = df["Temperature"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df["Temperature"] < lower_bound) | (df["Temperature"] > upper_bound)]

# Count the number of outliers
num_outliers = outliers.shape[0]

print(f"Number of outliers in Temperature: {num_outliers}")
```
