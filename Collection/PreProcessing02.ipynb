# **Remove Outliers**

```python
# Temperature
Q1 = df["Temp(C)"].quantile(0.25)
Q3 = df["Temp(C)"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

for i in range (100):
  outliers =(df['Temp(C)']<lower_bound) | (df['Temp(C)']>upper_bound)

  if not outliers.any():
    print(f"break loop")
    break

df["Temp(C)"]=df['Temp(C)'].mask(outliers,df['Temp(C)'].rolling(window=10,center=True,min_periods=1).mean())
```

```python
#Temperature outlier check after removal

Q1 = df["Temp(C)"].quantile(0.25)
Q3 = df["Temp(C)"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df["Temp(C)"] < lower_bound) | (df["Temp(C)"] > upper_bound)]
num_outliers = outliers.shape[0]
print(f"Number of outliers in Temp(C): {num_outliers}")
```

```python
#save dataset
df.to_csv('energy_iter13.csv',index=False)
print("saved with clean data")
```

```python
# Outliers removal for Generation

Q1 = df["Generation(MW)"].quantile(0.25)
Q3 = df["Generation(MW)"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

for i in range (100):
  outliers =(df['Generation(MW)']<lower_bound) | (df['Generation(MW)']>upper_bound)

  if not outliers.any():
    print(f"break loop")
    break

df["Generation(MW)"]=df['Generation(MW)'].mask(outliers,df['Generation(MW)'].rolling(window=7,center=True,min_periods=1).mean())
```

```python
#Generation outliers check after removal

Q1 = df["Generation(MW)"].quantile(0.25)
Q3 = df["Generation(MW)"].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df["Generation(MW)"] < lower_bound) | (df["Generation(MW)"] > upper_bound)]
num_outliers = outliers.shape[0]
print(f"Number of outliers in Generation(MW):{num_outliers}")
```

```python
#save dataset
df.to_csv('energy_iter13.csv',index=False)
print("saved with clean data")
```

```python
#we dont have any outlier in Demand .So ,we don't need to perform outlier removal operation for this
#Day Peak Generation outlier check

Q1 = df["Demand(MW)"].quantile(0.25)
Q3 = df["Demand(MW)"].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df["Demand(MW)"] < lower_bound) | (df["Demand(MW)"] > upper_bound)]
num_outliers = outliers.shape[0]
print(f"Number of outliers in Demand(MW): {num_outliers}")
```

# **MinMax Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Convert selected features all values between 0 to 1

features=["Demand(MW)","Generation(MW)","Temp(C)","DemandGenGap(MW)"]
scaler=MinMaxScaler(feature_range=(0,1))
df[features]=scaler.fit_transform(df[features])

print("scaled done")
```

```python
# Save the scaler For future reverse transformation(Scaling values to original values)

import joblib
joblib.dump(scaler,"minmax_scaler2.pkl")
```
