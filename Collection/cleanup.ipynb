```python
import pandas as pd
import numpy as np
```

### Store the extracted data in a dataframe:

```python
df = pd.read_csv('extracted_energy_data.csv')
df
```

### Convert the data to a datetime format, then sort in descending order:

```python
df['Report_Date'] = pd.to_datetime(df['Report_Date'], format='%d.%m.%y')
df = df.sort_values(by='Report_Date', ascending=False).reset_index(drop=True)
df
```

### Drop unnecessary rows, change datetime format:

```python
df = df.drop(columns=['File_Name'])
df['Report_Date'] = df['Report_Date'].dt.strftime('%d-%m-%Y')
df
```

```python
df.to_csv('cleaned_energy_datewise.csv', index=False)
```
