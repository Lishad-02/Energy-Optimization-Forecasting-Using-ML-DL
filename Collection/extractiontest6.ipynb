```python

# Replace blank strings with NaN
df.replace(r'^\s*$', pd.NA, regex=True, inplace=True)

# Find columns with only 0 or NaN
zero_or_nan_cols = [
    col for col in df.columns
    if df[col].dropna().apply(lambda x: x == 0).all()
]

print("Columns with only 0s or NaNs (no real values):")
print(zero_or_nan_cols)
```
